```python
X_data = np.genfromtxt('X_data.csv', delimiter=',')
Y_data = np.genfromtxt('Y_data.csv', delimiter=',')
Y_data = Y_data.reshape(712,1)
Y_data = Y_data.T
X_dev_data = np.genfromtxt('X_dev_data.csv', delimiter=',')
Y_dev_data = np.genfromtxt('Y_dev_data.csv', delimiter=',')
Y_dev_data = Y_dev_data.reshape(179,1)
```

```python
layers_dims = [10, 8, 6, 4,2,1]
```

```python
def initialize_parameters_deep(layer_dims):
    
    np.random.seed(3)
    parameters = {}
    L = len(layer_dims) # number of layers in the network

    for l in range(1, L):
   
        
        parameters['W' + str(l)] = np.random.randn(layer_dims[l], layer_dims[l-1]) * 0.01
        parameters['b' + str(l)] = np.zeros((layer_dims[l],1))
      
      
        assert(parameters['W' + str(l)].shape == (layer_dims[l], layer_dims[l - 1]))
        assert(parameters['b' + str(l)].shape == (layer_dims[l], 1))

        
    return parameters
```

```python
#initialize parameter
np.random.seed(3)
W1 = np.random.rand(8, 10) * 0.01
b1 = np.zeros((8,1))
W2 = np.random.rand(1, 8) * 0.01
b2 = np.zeros((1,1))
```

```python
def forward_propagation(W,b,x):
    z = np.dot(W,x) + b
    return np.maximum(0,z),z
```

```python
def forward_propagation_L(W,b,x):
    z = np.dot(W,x) + b
    return 1 / (1 + np.exp(-z))
```

```python
def deriv_ReLU(Z):
    return Z>0
```

```python
learning_rate = .1
```

```python
def test(W1,W2,b1,b2,X,Y_test):
    a1,cashe1 = forward_propagation(W1,b1,X.T)
    a2 = forward_propagation_L(W2,b2,a1)
    Y_pred = []
    for i in range(a2.shape[1]):
        if a2[0][i]>0.5:
            Y_pred.append(1.0)
        else:
            Y_pred.append(0.0)
    
    correct = 0
    for i in range (len(Y_pred)):
        if Y_pred[i]==Y_test[i][0]:
            correct+=1
    
    return correct/len(Y_pred)*100
```

```python
for i in range(10000):
    a1,cache1 = forward_propagation(W1,b1,X_data.T)
    a2 = forward_propagation_L(W2,b2,a1)
    cost = (-1/712)*np.sum(Y_data*np.log(a2)+(1-Y_data)*np.log(1-a2))

    
    dz2 = (a2 - Y_data)
    dW2 = np.dot(dz2,a1.T)/712
    db2 = np.sum(dz2)/712

    dz1 =  W2.T.dot(dz2) * deriv_ReLU(cache1)
    dW1 = np.dot(dz1,X_data)/712
    db1 = np.sum(dz1)/712
    
    W2=W2 - learning_rate*dW2
    b2 = b2 - learning_rate*db2
    W1=W1 - learning_rate*dW1
    b1 = b1 - learning_rate*db1
    
    acc = test (W1,W2,b1,b2,X_dev_data,Y_dev_data)
    print("Iteration",i,"--Cost:",cost,"Accuracy:",acc)

    
```
